## Pitch Class and Octaves

```python
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd
```

```python
def generate_sinusoid_pitches(pitches=[69], duration=0.5, Fs=4000, amp=1):
    N = int(duration * Fs)
    t = np.arange(N) / Fs
    x = []
    for p in pitches:
        freq = 2 ** ((p -69) / 12) * 440
        x = np.append(x, np.sin(2 * np.pi * freq * t))
    x = amp * x / np.max(x)
    return x, t
```

```python
# 途中経過の確認
duration = 0.5
Fs = 22050
N = int(duration * Fs)
t = np.arange(N) / Fs
x = []
x = np.append(x, np.sin(2 * np.pi * 50 * t))

plt.plot(t,x)

```

```python
duration = 1
Fs = 22050

pitches = [36,48,60,72,84,96,108]
x, t = generate_sinusoid_pitches(pitches=pitches, duration=duration, Fs=Fs, amp=0.5)
ipd.display(ipd.Audio(data=x, rate=Fs))
```

## Musical Scales

```python
duration = 0.5
Fs = 22050

x_maj, t = generate_sinusoid_pitches(
    pitches=[60,62,64,65,67,69,71,72], duration=duration, Fs=Fs, amp=0.5
)

x_min, t = generate_sinusoid_pitches(
    pitches=[60, 62, 63, 65, 67, 68, 70, 72], duration=duration, Fs=Fs, amp=0.5
)

ipd.display(ipd.Audio(data=x_maj, rate=Fs))
ipd.display(ipd.Audio(data=x_min, rate=Fs))
```

```python
duration = 0.5
Fs = 22050

x_chromatic, t = generate_sinusoid_pitches(
    pitches=range(60,72), duration=duration, Fs=Fs, amp=0.5
)

ipd.display(ipd.Audio(data=x_chromatic, rate=Fs))
```
