## Pitch and Center Frequencies

F pitch(p) = 2(p-69)/12 * 440


```python
def mtof(note):
    freq = 2 ** ((note - 69) / 12) * 440
    return freq

def mton(note):
    chromatic = ['A ', 'A#', 'B ', 'C ', 'C#', 'D ', 'D#',  'E ', 'F ', 'F#', 'G ', 'G#']
    return chromatic[(note - 69) % 12]

for note in range(21, 109):
    print('Note = %3d(%2s%1d), Freq = %7.2f' % (note, mton(note), note//12-1, mtof(note)))

```
