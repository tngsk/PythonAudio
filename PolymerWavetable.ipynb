# Waveteble for Bitwig Polymer

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import soundfile as sf
```

```python
Fs = 44100
samples = 256
Freq = Fs / samples
t = np.arange(samples) / Fs
x = []

#1 Sin
x = np.append(x, signal.chirp(t, Freq, Freq, samples, method='linear', phi=-90))

#2 Triangle
x = np.append(x, signal.sawtooth(2 * np.pi * Freq * t + np.pi/2, width=0.5))

#3 Square
x = np.append(x, signal.square(2 * np.pi * Freq * t, duty=0.5))

#4 Sawtooth
x = np.append(x, signal.sawtooth(2 * np.pi * Freq * t, width=1))

# Visualization
plt.plot(np.arange(x.size), x)
plt.grid(which='both', axis='both')
plt.xlabel('Samples')
plt.ylabel('Amp')
plt.xlim(0, x.size)
plt.xticks(np.arange(0, x.size, 128))

# Write wavetable
out = 'Audio/Scipy.Signal.wav'
x = x.astype(np.float32)
sf.write(out, x, Fs)
```
