# Waveteble for Bitwig Polymer

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import soundfile as sf
```

```python
Fs = 44100
samples = 256
Freq = Fs / samples
t = np.arange(samples) / Fs
x = []

#1 Sin
x = np.append(x, signal.chirp(t, Freq, Freq, samples, method='linear', phi=-90))

#2 Triangle
x = np.append(x, signal.sawtooth(2 * np.pi * Freq * t + np.pi/2, width=0.5))

#3 Square
x = np.append(x, signal.square(2 * np.pi * Freq * t, duty=0.5))

#4 Sawtooth
x = np.append(x, signal.sawtooth(2 * np.pi * Freq * t, width=1))

# Visualization
plt.plot(np.arange(x.size), x)
plt.grid(which='both', axis='both')
plt.xlabel('Samples')
plt.ylabel('Amp')
plt.xlim(0, x.size)
plt.xticks(np.arange(0, x.size, 128))

# Write wavetable
out = 'Audio/Scipy.Signal.wav'
x = x.astype(np.float32)
sf.write(out, x, Fs)
```

```python
Fs = 44100
samples = 256
Freq = Fs / samples
t = np.arange(samples) / Fs
x = []

# Opsix ADTV

# SAW3
# Sawtooth
w1 = signal.sawtooth(2 * np.pi * Freq * 1 * t, width=1)
w2 = signal.sawtooth(2 * np.pi * Freq * 2 * t, width=1)
w3 = signal.sawtooth(2 * np.pi * Freq * 3 * t, width=1)
saw3 = w1 * (1.0/3.0) + w2 * (1.0/3.0) + w3 * (1.0/3.0)
x = np.append(x, saw3)

# SQR3
# Square
w1 = signal.square(2 * np.pi * Freq * 1 * t, duty=0.5)
w2 = signal.square(2 * np.pi * Freq * 2 * t, duty=0.5)
w3 = signal.square(2 * np.pi * Freq * 3 * t, duty=0.5)
sqr3 = w1 * (1.0/3.0) + w2 * (1.0/3.0) + w3 * (1.0/3.0)
x = np.append(x, sqr3)

# TRI3
# Triangle

w1 = signal.sawtooth(2 * np.pi * Freq * 1 * t + np.pi/2, width=0.5)
w2 = signal.sawtooth(2 * np.pi * Freq * 2 * t + np.pi/2, width=0.5)
w3 = signal.sawtooth(2 * np.pi * Freq * 3 * t + np.pi/2, width=0.5)
tri3 = w1 * (1.0/3.0) + w2 * (1.0/3.0) + w3 * (1.0/3.0)
x = np.append(x, tri3)

# Visualization
plt.plot(np.arange(x.size), x)
plt.grid(which='both', axis='both')
plt.xlabel('Samples')
plt.ylabel('Amp')
plt.xlim(0, x.size)
plt.xticks(np.arange(0, x.size, 128))

# Write wavetable
out = 'Audio/Opsix.ADTV.SAW3-SQR3-TRI3.wav'
x = x.astype(np.float32)
sf.write(out, x, Fs)
```
